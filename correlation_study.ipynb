```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
deterioration_data = pd.read_csv('/Users/banna/Desktop/CEE 4803/Project/Correlation Study/deterioration_rates.csv')

# Create a 1x3 subplot grid
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot with trendline for ESALS vs. Deterioration Rate
sns.regplot(data=deterioration_data, x='ESALS', y='Deterioration_Rate_Moving_Avg', scatter_kws={'alpha': 0.5}, ax=axes[0])
axes[0].set_title('ESALS vs. Deterioration Rate')
axes[0].set_xlabel('ESALS')
axes[0].set_ylabel('Deterioration Rate')

# Scatter plot with trendline for MAXSPEED vs. Deterioration Rate
sns.regplot(data=deterioration_data, x='MAXSPEED', y='Deterioration_Rate_Moving_Avg', scatter_kws={'alpha': 0.5}, ax=axes[1])
axes[1].set_title('MAXSPEED vs. Deterioration Rate')
axes[1].set_xlabel('MAXSPEED')
axes[1].set_ylabel('Deterioration Rate')

# Scatter plot with trendline for CYCLE NO. vs. Deterioration Rate
sns.regplot(data=deterioration_data, x='CYCLE', y='Deterioration_Rate_Moving_Avg', scatter_kws={'alpha': 0.5}, ax=axes[2])
axes[2].set_title('CYCLE NO. vs. Deterioration Rate')
axes[2].set_xlabel('CYCLE NO.')
axes[2].set_ylabel('Deterioration Rate')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
